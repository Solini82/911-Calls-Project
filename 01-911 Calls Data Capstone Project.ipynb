** In the titles column there are "Reasons/Departments" specified before the title code. These are EMS, Fire, and Traffic. Use .apply() with a custom lambda expression to create a new column called "Reason" that contains this string value.**

** Now let us begin to focus on time information. What is the data type of the objects in the timeStamp column? **

```python
df['timeStamp'].dtype
```

** You should have seen that these timestamps are still strings. Use [pd.to_datetime](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.to_datetime.html) to convert the column from strings to DateTime objects. **

```python
df['timeStamp'] = pd.to_datetime(df['timeStamp'])
```

** You can now grab specific attributes from a Datetime object by calling them. For example:**

    time = df['timeStamp'].iloc[0]
    time.hour

**You can use Jupyter's tab method to explore the various attributes you can call. Now that the timestamp column are actually DateTime objects, use .apply() to create 3 new columns called Hour, Month, and Day of Week. You will create these columns based off of the timeStamp column, reference the solutions if you get stuck on this step.**

```python
df['Hour'] = df['timeStamp'].apply(lambda x: x.hour)
df['Month'] = df['timeStamp'].apply(lambda x: x.month)
df['Day of Week'] = df['timeStamp'].apply(lambda x: x.dayofweek)
```

** Notice how the Day of Week is an integer 0-6. Use the .map() with this dictionary to map the actual string names to the day of the week: **

    dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}

```python
# Dictionary to map integers to day names
dmap = {0: 'Mon', 1: 'Tue', 2: 'Wed', 3: 'Thu', 4: 'Fri', 5: 'Sat', 6: 'Sun'}
```

```python
# Apply the mapping to the 'Day of Week' column
df['Day of Week'] = df['Day of Week'].map(dmap)
```

** Now use seaborn to create a countplot of the Day of Week column with the hue based off of the Reason column. **

```python
# Create a countplot
plt.figure(figsize=(10, 6))
sns.countplot(x='Day of Week', hue='Reason', data=df, palette='viridis')
# Add title and labels
plt.title('911 Calls by Day of Week and Reason')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Calls')
```

** Now create a gropuby object called byMonth, where you group the DataFrame by the month column and use the count() method for aggregation. Use the head() method on this returned DataFrame. **

```python
# Grouping the DataFrame by the 'Month' column and counting occurrences
byMonth = df.groupby('Month').count()
# Display the first few rows of the resulting DataFrame
byMonth.head()
```

** Now create a simple plot off of the dataframe indicating the count of calls per month. **

```python
# Plotting the count of calls per month
plt.figure(figsize=(10, 6))
byMonth['lat'].plot()
# Add title and labels
plt.title('Count of 911 Calls per Month')
plt.xlabel('Month')
plt.ylabel('Number of Calls')
```

** Now see if you can use seaborn's lmplot() to create a linear fit on the number of calls per month. Keep in mind you may need to reset the index to a column. **

```python
# Reset the index to make 'Month' a column
byMonth = byMonth.reset_index()
# Use Seaborn's lmplot to create a linear fit
sns.lmplot(x='Month', y='lat', data=byMonth)
# Add title and labels
plt.title('Linear Fit of 911 Calls per Month')
plt.xlabel('Month')
plt.ylabel('Number of Calls')
```

**Create a new column called 'Date' that contains the date from the timeStamp column. You'll need to use apply along with the .date() method. ** 

```python
df['Date'] = df['timeStamp'].apply(lambda x: x.date())
```

** Now groupby this Date column with the count() aggregate and create a plot of counts of 911 calls.**

```python
# Group by the 'Date' column and count occurrences
byDate = df.groupby('Date').count()
# Plot the counts of 911 calls per date
plt.figure(figsize=(12, 6))
byDate['lat'].plot()
# Add title and labels
plt.title('Count of 911 Calls per Date')
plt.xlabel('Date')
plt.ylabel('Number of Calls')
```

** Now recreate this plot but create 3 separate plots with each plot representing a Reason for the 911 call**

```python
# First, extract the reason from the 'title' column, assuming 'title' has a format like 'EMS: BACK PAINS'
df['Reason'] = df['title'].apply(lambda x: x.split(':')[0])
# Now group by 'Date' and count occurrences for each reason
byDate_EMS = df[df['Reason'] == 'EMS'].groupby('Date').count()
byDate_Fire = df[df['Reason'] == 'Fire'].groupby('Date').count()
byDate_Traffic = df[df['Reason'] == 'Traffic'].groupby('Date').count()
# Plotting each reason separately
plt.figure(figsize=(12, 15))
# EMS plot
plt.subplot(3, 1, 1)
byDate_EMS['lat'].plot()
plt.title('Count of 911 Calls per Date for EMS')
plt.xlabel('')
plt.ylabel('Number of Calls')
# Fire plot
plt.subplot(3, 1, 2)
byDate_Fire['lat'].plot()
plt.title('Count of 911 Calls per Date for Fire')
plt.xlabel('')
plt.ylabel('Number of Calls')
# Traffic plot
plt.subplot(3, 1, 3)
byDate_Traffic['lat'].plot()
plt.title('Count of 911 Calls per Date for Traffic')
plt.xlabel('Date')
plt.ylabel('Number of Calls')
# Adjust layout
plt.tight_layout()
```

** Now let's move on to creating  heatmaps with seaborn and our data. We'll first need to restructure the dataframe so that the columns become the Hours and the Index becomes the Day of the Week. There are lots of ways to do this, but I would recommend trying to combine groupby with an [unstack](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.unstack.html) method. Reference the solutions if you get stuck on this!**

```python
# Group by 'Day of Week' and 'Hour' and count the number of calls
dayHour = df.groupby(by=['Day of Week', 'Hour']).count()
# Restructure the DataFrame with unstack so that 'Hour' becomes the columns
dayHour = dayHour['lat'].unstack()
```

** Now create a HeatMap using this new DataFrame. **

```python
plt.figure(figsize=(12, 6))
# Create the heatmap
sns.heatmap(dayHour, cmap='viridis', annot=True, linewidths=.5)
# Add title and labels
plt.title('Heatmap of 911 Calls by Day of Week and Hour')
plt.xlabel('Hour')
plt.ylabel('Day of Week')
```

** Now create a clustermap using this DataFrame. **

```python
# Create the clustermap
sns.clustermap(dayHour, cmap='viridis', standard_scale=1, linewidths=.5)
# Add title (clustermap automatically includes axes labels)
plt.title('Clustermap of 911 Calls by Day of Week and Hour', pad=70)
```
